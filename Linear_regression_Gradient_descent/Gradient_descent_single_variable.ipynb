```python
#Importing Libraries
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import sympy as sym
```

```python
initial_point = 3
rate = 0.1
iterations = 0
max_iterations = 1000 
precision = 0.000001 
step_size = 1 
x = np.random.randint(1,10,20)
fx = x**2
df = lambda x: 2*x
```

```python
while step_size > precision and iterations < max_iterations:
    previous_x = initial_point 
    initial_point = initial_point - rate * df(previous_x) 
    step_size = abs(initial_point - previous_x) 
    iterations = iterations + 1 
    print("Iteration",iterations,"\nX value is",initial_point) 

print("The local minimum of function is at", initial_point)
```
