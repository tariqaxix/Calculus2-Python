```python
#Importing necessary libraries
from sympy import*
import matplotlib.pyplot as plt
%matplotlib inline
x = symbols('x')
```

```python
#Taking input from the user
funct_1 = input("Enter the function: ")
a = float(input("Enter a: "))
value = float(input("Number of evaluation: "))
```

```python
f = lambdify (x, funct_1)
f_of_a = f(a)
diff = Derivative(funct_1, x).doit()
f_der = lambdify (x, diff)
f_der_of_x = f_der(a)
linear_expression = f_of_a + (f_der_of_x * (x-a))
linear_function = lambdify(x, linear_expression)
result = linear_function(value)
print(result)
```

```python
#Plotting the graph
a = [i for i in range(0,30)]
y = [f(i) for i in a]
y_new = [linear_function(i) for i in a]
plt.figure(figsize = (10,10))
plt.plot(a, y, label = funct_1)
plt.plot(a, y_new, label = linear_expression)

plt.show()
```
