```python
#Importing necessay Libraries
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sympy.abc import x
from sympy import *
x = sym.symbols('x')
```

```python
#Taking input from the user
funct_2 = input("Enter a function: ")
x_init = int(input("Enter x initial: "))
```

```python
#Code and plotting graph
f = sym.lambdify(x, funct_2)
x_axis = np.linspace(-1, 5, 20)
plt.figure(figsize = (20,10))
plt.plot(x_axis, f(x_axis), linewidth = 1)

diff = Derivative(funct_2, x).doit()
while abs(f(x_init))>=0.00001:
    x1 = x_init - f(x_init)/diff.subs(x, x_init)
    plt.plot((x1, x_init),(0, f(x_init)), 'r--')
    x_init = x1
plt.xlim(-1,5)
plt.ylim(0,20)
print(diff) 
print(x1)
```
