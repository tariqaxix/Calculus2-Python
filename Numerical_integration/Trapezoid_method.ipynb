```python
#Importing necessary libraries
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sympy.abc import x
from sympy import *
```

```python
#Taking input from user
func = input("Enter an function: ")
initial_point = int(input("Enter starting point: "))
final_point = int(input("Enter ending point: "))
num_of_int = int(input("Number of intervals: "))
```

```python
#Code for trapezoid method
x = sym.symbols('x')
f = sym.lambdify(x, func)
dx = (final_point - initial_point)/num_of_int
A = 1/2 *(f(initial_point) + f(final_point))
for i in range(1, num_of_int):
    A = A + f(initial_point + i*dx)
Area = dx * A
print("The area under the curve using Trapezoid method is", Area)
```

```python
#Graph for Trapezoid method
X = np.linspace(initial_point,final_point,100)
x = np.linspace(initial_point,final_point,num_of_int+1)
Y = f(X)
plt.figure(figsize = (15,10))
plt.plot(X,Y, color='black', linewidth=2, markersize=50)

for i in range(num_of_int):
    initial_point = [x[i],x[i],x[i+1],x[i+1]]
    final_point = [0,f(x[i]),f(x[i+1]),0]
    plt.fill(initial_point,final_point)
plt.title('Area under the curve, Trapezoid method')
plt.show()
```
