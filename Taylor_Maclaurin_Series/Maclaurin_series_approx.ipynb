```python
#Importing Libraries
import sympy as sy
import numpy as np
from sympy.functions import *
import matplotlib.pyplot as plt
from sympy.plotting import plot
x = sy.Symbol('x')
```

```python
def factorial(n):
    if n <= 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
def maclaurin(func, n):
    i = 0
    function = 0
    while i <= n:
        function = function + (func.diff(x, i).subs(x, 0)/(factorial(i)))*x**i
        i += 1
    return function
```

```python
#Function to be evaluated
func = maclaurin(sin(x), 9)
print(func)
print('Value is', maclaurin(sin(x), 9).subs(x,1))
```

```python
#Plotting the graph
p = plot(sin(x),maclaurin(sin(x),9),
         (x,-5,5),legend=True, show=False)
p[0].line_color = 'blue'
p[1].line_color = 'green'
p.title = 'Maclaurin Series Expansion for Sine'
p.show()
```
