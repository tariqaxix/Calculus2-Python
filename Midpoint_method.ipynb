```python
import sympy as sym
from sympy.abc import x
from scipy.integrate import quad
from math import sin, cos, tan, log, e, pi
import numpy as np
from random import randint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
func = input('Enter the function: ')
f = sym.lambdify(x, func)
initial_point = int(input('Enter the start of the interval: '))
final_point = int(input('Enter the end of the interval: '))
```

```python
def midpoint_method(f, initial_point, final_point):
    n = randint(5, 50)
    x_val = np.linspace(initial_point, final_point, n+1)
    dx = (final_point-initial_point)/n
    area = 0

    array = np.linspace(initial_point, final_point, 50)
    plt.plot(array, f(array), color='black')
    for i in range(len(x_val)-1):
        area += dx*f((x_val[i]+x_val[i+1])/2) 

        
        array = np.linspace(x_val[i], x_val[i+1], 50)
        plt.fill_between(array, f((x_val[i]+x_val[i+1])/2))
        
    return(area)
midpoint_method(f, initial_point, final_point)
```
