```python
#Importing Libraries
import sympy as sy
import numpy as np
from sympy.functions import *
import matplotlib.pyplot as plt
from sympy.plotting import plot
x = sy.Symbol('x')
  
```

```python
def factorial(n):
    if n <= 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
def taylor(func, c, n):
    i = 0
    function = 0
    while i <= n:
        function = function + (func.diff(x, i).subs(x, c)/(factorial(i)))*(x - c)**i
        i += 1
    return function
```

```python
#Function to be evaluated
func = taylor(cos(x), 1, 5)
print(func)
print('Value is', taylor(cos(x),2,5).subs(x,1))
```

```python
#Plotting the graph
p = plot(cos(x),taylor(cos(x),1,5),
         (x,-5,5),legend=True, show=False)
p[0].line_color = 'blue'
p[1].line_color = 'green'
p.title = 'Taylor Series Expansion'
p.show()
```
